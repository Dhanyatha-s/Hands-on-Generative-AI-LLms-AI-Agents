```python
from transformers import pipeline
```

```python
# 1.Initilaize the text generation pipeline
generator = pipeline("text-generation", model = "distilgpt2")
```

```python
# 2. Define the Prompt
prompt = "The AI has taken over the job market and most the jokseekers are now"
```

```python
# 3. Generate the Text with sampling parameters
output = generator(
    prompt,
    max_length = 100, #stopes after 100 tokens
    num_return_sequences = 1, # generate 1 sequence of word
    do_sample = True, # Crutial enable sampling for non-greedy generation
    top_k = 50, # top most likely next word
    top_p = .95, #Nucleus Sampling threshold
    temperature = .75, # controls randomness: higher - crazier, lower - safer
)
```
